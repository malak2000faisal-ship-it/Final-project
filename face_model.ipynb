```python
import os, random
import numpy as np
from PIL import Image

from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.preprocessing import LabelEncoder

import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Flatten, Dense, Dropout
```

```python
STUDENTS_FOLDER = 'Students_Images'  
IMG_SIZE        = (128, 128)         
SEED            = 42             

random.seed(SEED)
np.random.seed(SEED)
tf.keras.utils.set_random_seed(SEED)
```

```python
STUDENTS_FOLDER
```

```python
def load_img(path):
    img = Image.open(path).convert('RGB').resize(IMG_SIZE)
    return np.array(img).astype(np.float32) / 255.0  

X      = []  
labels = []  

for person_name in os.listdir(STUDENTS_FOLDER):
    person_folder = os.path.join(STUDENTS_FOLDER, person_name)
    
    if os.path.isdir(person_folder):
        for filename in os.listdir(person_folder):
            if filename.lower().endswith(('.jpg', '.jpeg', '.png')):
                path = os.path.join(person_folder, filename)
                X.append(load_img(path))
                labels.append(person_name)

for filename in os.listdir(STUDENTS_FOLDER):
    if filename.lower().endswith(('.jpg', '.jpeg', '.png')):
        path = os.path.join(STUDENTS_FOLDER, filename)
        
        name = os.path.splitext(filename)[0] 
        name = '_'.join(name.split('_')[:-1])
        
        X.append(load_img(path))
        labels.append(name)

X = np.array(X)
print('Dataset shape:', X.shape)
print('Students:', labels)
```

```python
le = LabelEncoder()
y  = le.fit_transform(labels)
n_class = len(le.classes_)
```

```python
n_class
```

```python
le.classes_
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.15, random_state=SEED
)
```

```python
X_train, X_val, y_train, y_val = train_test_split(
    X_train, y_train, test_size=0.15, random_state=SEED
)
```

```python
inp = Input(shape=(128, 128, 3))
inp
```

```python
inp = Input(shape=(IMG_SIZE[0], IMG_SIZE[1], 3))
x = Conv2D(32, 3, activation='relu', padding='same')(inp)
x = MaxPooling2D()(x)
x = Conv2D(64, 3, activation='relu', padding='same')(x)
x = MaxPooling2D()(x)
X
```

```python
x = Conv2D(128, 3, activation='relu', padding='same')(x)
x = MaxPooling2D()(x)
x = Flatten()(x)
X
```

```python
x = Dropout(0.3)(x)
x = Dense(128, activation='relu')(x)
out = Dense(n_class, activation='softmax')(x)
```

```python
model = Model(inp, out)
model.compile(
    optimizer = 'adam',
    loss      = 'sparse_categorical_crossentropy',
    metrics   = ['accuracy']
)

model.summary()
```

```python
history = model.fit(
    X_train, y_train,
    epochs          = 50,
    batch_size      = 8,
    validation_data = (X_val, y_val)
)
```
